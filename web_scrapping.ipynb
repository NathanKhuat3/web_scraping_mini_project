```python
from bs4 import BeautifulSoup
import requests
import pandas as pd

url = r"https://en.wikipedia.org/wiki/List_of_comedy_anime"

website = requests.get(url)
raw_code = BeautifulSoup(website.text, 'html')
print(raw_code)
```

```python
main_data = raw_code.find("table", class_ = "wikitable sortable")
# location of the table: <table class="wikitable sortable jquery-tablesorter" border="1">
```

```python
# Get the main titles of the desired table 
raw_titles = main_data.find_all('th')
titles = [title.text.strip() for title in raw_titles]
titles
```

```python
# Create the data frame
main_table = pd.DataFrame(columns = titles)
main_table
```

```python
raw_rows = main_data.find_all('tr')
for row in raw_rows[1:]:
    row_data = row.find_all('td')
    clean_row_data = [clean_row.text.strip() for clean_row in row_data]
    df_length = len(main_table)
    # Here, I found that, for some reason, the rows  that had no value in the column Ref sometimes are not registered as '' like normal
    # and does not have data at all in that space, this code here fixed it
    if len(clean_row_data) < len(titles):
        clean_row_data.append('')
    main_table.loc[df_length] = clean_row_data
main_table
```
